## Function

### creating functions

```python
#find even or add
def even_odd(x):
    if x%2==0:
        print("even no")
    else:
        print("odd")
```

```python
even_odd(10)
```

```python
#find factorial of no
def factorial(a):
    fact=1
    for i in range(1,a+1):
        fact=fact*i
    return fact
    
```

```python
factorial(6)
```

```python
#sum of all natural numbers
def sum_of_natural_numbers(n):
    sum=0
    for i in range(1,n+1):
        sum+=i
    return  sum
    
```

```python
sum_of_natural_numbers(5)
```

### lambda function

```python
l_var=lambda a:a+10
l_var(3)
```

#### check even or odd using lambda function

```python
even_odd=lambda x:'even' if x%2==0 else 'odd'
```

```python
even_odd(4)
```

#### add two integers

```python
add=lambda a,b:a+b
add(10,4)
```
