### set operations

```python
#union operation 
set1={1,2,3,4,5}
set2={3,4,5,6,7}
set1.union(set2)

```

```python
set1|set2
```

```python
#intersection operation
set1&set2
```

```python
set1.intersection(set2)
```

```python
#difference
set1-set2
```

```python
set1.difference(set2)
```

```python
#set symmetric difference
set1^set2
```
